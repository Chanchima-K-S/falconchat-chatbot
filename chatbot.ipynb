```python
import torch
from transformers import AutoTokenizer, pipeline
```

```python
model = "tiiuae/falcon-7b-instruct"
```

```python
tokenizer = AutoTokenizer.from_pretrained(model)
```

```python
pipeline = pipeline(
    "text-generation",
    model=model,
    tokenizer=tokenizer,
    torch_dtype=torch.bfloat16,
    trust_remote_code=True,
    max_length = 200,
    device_map="auto"
)
```

```python
prompt = "What is a blackhole?"
```

```python
newline_token = tokenizer.encode("\n")[0]

my_name = "Chanchima"
your_name = "Hemanth"
dialogue = []

while True:
  user_input = input("> ")
  dialogue.append(f"{my_name}: {user_input}")
  prompt = "\n".join(dialogue) + f"\n{your_name}: "
  sequences = pipeline(
      prompt,
      do_sample=True,
      max_length = 500,
      top_k=10,
      return_full_text = False,
      num_return_sequences=1,
      eos_token_id=newline_token,
      pad_token_id=tokenizer.eos_token_id,
  )
  print(sequences[0]["generated_text"])
  dialogue.append(f"{sequences[0]['generated_text']}")
```
